```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
```

```python
# features with nan values
features_with_na
```

```python
dataset['Electrical'].isnull().mean()
```

```python
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean()*100,4),'% missing values')
```

```python
## Let's plot some diagram

for feature in features_with_na:
    data = dataset.copy()
    # let's make a variable that indicate 1 if the observation was missing or 0 otherwise
    data[feature] = np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
## list of numerical variables

dataset['SalePrice'].dtypes != 'object'
```

```python
numerical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes!='O']
```

```python
print(len(numerical_feature))
dataset[numerical_feature].head()
```

```python
dataset.info()
```

```python
## temporal variables(eg: datetime variable)
year_feature = [feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year sold')
```

```python
## here we will compare the difference between all years features with saleprice
data = dataset.copy()
for feature in year_feature:
    if feature != 'YrSold':
        data[feature] = data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
## Numerical Variables 52 types

# Continuous variable and discreate varible

discreate_feature = [feature for feature in numerical_feature if len(dataset[feature].unique())<=25]
len(discreate_feature)
```

```python
dataset[discreate_feature].head()
```

```python
## lets find the relationship between discreate and saleprice

data = dataset.copy()
for feature in discreate_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
## Continuous variable

continuous_feature = [feature for feature in numerical_feature if feature not in discreate_feature]
len(continuous_feature)
```

```python
### lets 

data = dataset.copy()
for feature in continuous_feature:
    data.groupby
```

```python
data = dataset.copy()
for feature in continuous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
### EDA Part 2
```

```python
data=dataset.copy()

for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

```python
## outliers

for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
