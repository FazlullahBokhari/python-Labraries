```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
# 1. Find out how many male and female passenger was onboarded .

total_number_of_female = len(data[data['Sex'] == 'female'])
print("total_number_of_female passenger was onboard = ",total_number_of_female)
total_number_of_male = len(data[data['Sex'] == 'male'])
print("total_number_of_male passenger was onboard = ",total_number_of_male)
```

```python
# 2. how many survived we have .
total_number_of_survived = len(data[data['Survived'] == 1])
print("total_number_of_survived = ",total_number_of_survived)
```

```python
# 3. how many casuality we have
total_number_of_casuality = len(data[data['Survived'] == 0])
print("total_number_of_casuality = ",total_number_of_casuality)
```

```python
# 4. what is name of a person who is the eldest one .
data[data["Age"] == max(data["Age"])]['Name']
```

```python
# 5. how many passenger do we have in first , second and third class 
total_number_of_passenger_in_first_class = len(data[data["Pclass"] == 1])
print("total_number_of_passenger_in_first_class = ",total_number_of_passenger_in_first_class)

total_number_of_passenger_in_second_class = len(data[data["Pclass"] == 2])
print("total_number_of_passenger_in_second_class = ",total_number_of_passenger_in_second_class)

total_number_of_passenger_in_third_class = len(data[data["Pclass"] == 3])
print("total_number_of_passenger_in_third_class = ",total_number_of_passenger_in_third_class)
```

```python
# 6. how many person we have whose name starts with "s"
def get_count(data):
    count = 0
    for name in data["Name"]:
        if name[0] == "S":
            count +=1  
    return count       

get_count(data)
```
