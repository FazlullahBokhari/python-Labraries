```python
# 4 . in which month we have trageted most of the customer . 
def get_month_count(data):
    month_count = {}
    for m in data['month']:
        if m in month_count:
            month_count[m] = month_count[m] + 1
        else:
            month_count[m] = 1
    print(month_count)
    
    frequency = 0
    month = ""
    for mo,count in month_count.items():
        if count>frequency:
            frequency = count
            month = mo
    return month        

get_month_count(data)
```

```python
# 5 . which mode of call is giving you more result
def get_contact_success_mode(data):
    count = {}
    contact_modes = list(set(data['contact']))
    for mode in contact_modes:
        a = len(data[(data['contact']==mode) & (data['poutcome']=="success")])
        if mode not in count:
            count[mode] = a
    print(count) 
    frequency = 0
    contact_type = ""
    for mode,count in count.items():
        if count>frequency:
            frequency = count
            contact_type = mode
    return contact_type, frequency        
        
get_contact_success_mode(data)        
```
