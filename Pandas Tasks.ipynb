```python
# 6. how many person we have whose name starts with "s"
def get_count(data):
    count = 0
    for name in data["Name"]:
        if name[0] == "S":
            count +=1  
    return count       

get_count(data)
```

```python
# 4 . in which month we have trageted most of the customer . 
def get_month_count(data):
    month_count = {}
    for m in data['month']:
        if m in month_count:
            month_count[m] = month_count[m] + 1
        else:
            month_count[m] = 1
    print(month_count)
    
    frequency = 0
    month = ""
    for mo,count in month_count.items():
        if count>frequency:
            frequency = count
            month = mo
    return month        

get_month_count(data)
```
